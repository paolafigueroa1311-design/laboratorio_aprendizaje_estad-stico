```python
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
data=pd.read_excel("Boston Housing Dataset 1978.xlsx")
data.head(n=5)
```

```python
data.tail(n=5)
```

```python
data.describe()
```

```python
y= data.MEDV
x1=data.ZN.values.reshape([-1,1])
x2= data.INDUS.values.reshape([-1,1])
x3= data.CHAS.values.reshape([-1,1])
x4= data.NOX.values.reshape([-1,1])
x5= data.RM.values.reshape([-1,1])
x6= data.AGE.values.reshape([-1,1])
x7= data.DIS.values.reshape([-1,1])
x8= data.RAD.values.reshape([-1,1])
x9= data.TAX.values.reshape([-1,1])
x10= data.PTRATIO.values.reshape([-1,1])
x11= data.B.values.reshape([-1,1])
x12= data.LSTAT.values.reshape([-1,1])
x13=data.CRIM.values.reshape([-1,1])
n=len(y)
unos=np.ones([n,1])
X=np.hstack([unos, x1,x2,x3,x4,x5,x6,x7,x8,x9,x10,x11,x12,x13])
# X es la concatenación de todos los factores
ols= sm.OLS(y,X)
results= ols.fit()
results.summary()
```
