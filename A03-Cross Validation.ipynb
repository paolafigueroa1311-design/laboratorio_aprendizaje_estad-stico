```python
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
data=pd.read_excel("Boston Housing Dataset 1978.xlsx")
data.head(n=5)
data
```

```python
data2=data.dropna()
data2
```

```python
data3=data2.drop_duplicates()
data3
```

```python
n=len(data3)
train= np.random.choice(n, 226, False)
xtrain=data3.iloc[train]

test=np.setdiff1d(np.arange(n),train)
xtest=data3.iloc[test]
xtrain, xtest
```

```python
cols = ["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE",
        "DIS","RAD","TAX","PTRATIO","B","LSTAT"]

x_train_list = []
x_test_list = []

for col in cols:
    x_train_list.append(xtrain[col].values.reshape(-1, 1))
    x_test_list.append(xtest[col].values.reshape(-1, 1))

X_train = np.hstack(x_train_list)
X_test = np.hstack(x_test_list)
y_train = xtrain["MEDV"].values.reshape(-1, 1)
y_test = xtest["MEDV"].values.reshape(-1, 1)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
scaler=StandardScaler().fit(X_train)
xescalado_train=scaler.transform(X_train)
lr=LinearRegression()
```

```python
lr.fit(xescalado_train, y_train)
```

```python
xescalado_test = scaler.transform(X_test)
y_test_predict = lr.predict(xescalado_test)
```

```python
y_test_predict - y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,lr.predict(xescalado_test))
```

```python
r2_score(y_train,lr.predict(xescalado_train))
```
