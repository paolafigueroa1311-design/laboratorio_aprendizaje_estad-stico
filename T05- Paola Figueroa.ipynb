### 1.1 Realiza una regresión tomando 'mpg' como salida y eliminando la columna 'model'. Considera todos los demás factores como numéricos/ordinales.

Calcula el R2 e interpreta los signos de los betas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from scipy import stats 
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
```

```python
data= pd.read_excel("/Users/paofigueroa/Documents/sem 5/Lab de aprendizaje estadístico/Motor Trend Car Road Tests.xlsx")
data = data.drop(columns="model")
data.head()
```

```python
data2=data.dropna().drop_duplicates()
data2
```

```python
cols=data2.columns
cols=cols.drop("mpg")
x_list = []

for col in cols:
    x_list.append(data2[col].values.reshape(-1, 1))


X = np.hstack(x_list)
y = data2["mpg"].values.reshape(-1, 1)

scaler=StandardScaler().fit(X)
Xescalada=scaler.transform(X)

x=np.ones((len(X),1))
X_e=np.hstack((x,Xescalada))

ols=sm.OLS(y,X_e)
results=ols.fit()
results.summary()
```

```python
X_mpg = data2.drop(columns=['mpg']).values
y_mpg = data2['mpg'].values
```

```python
X_mpg_train, X_mpg_test, y_mpg_train, y_mpg_test = train_test_split(X_mpg, y_mpg, train_size=0.4, random_state=137)
scaler = StandardScaler().fit(X_mpg_train)
X_mpg_train_scaled = scaler.transform(X_mpg_train)
X_mpg_test_scaled = scaler.transform(X_mpg_test)
```

```python
X_mpg_train_sm = sm.add_constant(X_mpg_train_scaled)
modelo_mpg = sm.OLS(y_mpg_train, X_mpg_train_sm).fit()
```

```python
modelo_mpg.summary()
```

Desgloce de los coeficientes:
- x1 - cyl (40.7571): Por cada cilindro adicional, el mpg aumenta en 40.76 unidades. 
- x2 - disp (-2.4635): Por cada pulgada cúbica adicional de desplazamiento, el mpg disminuye en 2.46 unidades. Motores más grandes consumen más combustible.
- x3 - hp (-10.1581): Por cada fuerza adicional, el mpg disminuye en 10.16 unidades. A mayor potencia, mayor consumo.
- x4 - drat (14.1135): Por cada unidad adicional en la relación del eje trasero, el mpg aumenta en 14.11 unidades. Una mejor relación puede mejorar eficiencia.
- x5 - wt (-0.8005): Por cada peso adicional, el mpg disminuye en 0.80 unidades, ya que los autos más pesados pesados consumen más combustible.
- x6 - qsec (1.5727): Por cada segundo adicional en tiempo, el mpg aumenta en 1.57 unidades. Los autos más lentos suelen ser más eficientes en combustible (gastan menos)
- x7 - vs (32.4926): Un mejor motor aumenta el combustible.
- x8 - am (36.1394): una mejor transmisión aumenta el  mpg en 36.14 unidades.
- x9 - gear (-38.3551): Por cada velocidad adicional, el mpg disminuye en 38.36 unidades.
- x10 - carb (11.9707): Por cada carburador adicional, el mpg aumenta en 11.97 unidades.

Podemos ver que los p-values son muy altos (mayores al 0.05), lo que significa que es pura chiripa, es decir, ninguna variable es estadísticamente significativa, lo que sugiere que ninguna de las variables tiene efecto fuerte sobre Y.
El modelo logra explicar el 99% de la variabilidad de mpg.

```python
# Realiza un train-test-split donde se use el 40% de los datos para entrenar. Calcula el R2 de entrenamiento y de prueba.
X_mpg_test_sm = sm.add_constant(X_mpg_test_scaled)
y_mpg_pred = modelo_mpg.predict(X_mpg_test_sm)
r2_mpg_train= modelo_mpg.rsquared
r2_mpg_test= r2_score(y_mpg_test, y_mpg_pred)
print(f"R2 de entrenamiento: {r2_mpg_train}")
print(f"R2 de prueba: {r2_mpg_test}")
```

```python
# Añade regularización L2 con un hiperparámetro lambda decidido por ti. Cambia este valor y compara con varios distintos los R2 de entrenamiento y de prueba.
lambdas = [0.01, 0.1, 1, 10, 100]
resultados_mpg_train={}
resultados_mpg_test={}
for l in lambdas:
    modelo_ridge = Ridge(alpha=l, fit_intercept=True)
    modelo_ridge.fit(X_mpg_train_scaled, y_mpg_train)
    y_mpg_train_ridge = modelo_ridge.predict(X_mpg_train_scaled)
    y_mpg_test_ridge= modelo_ridge.predict(X_mpg_test_scaled)
    # calculo de r2
    r2_train_ridge = r2_score(y_mpg_train, y_mpg_train_ridge)
    r2_test_ridge = r2_score(y_mpg_test, y_mpg_test_ridge)
    resultados_mpg_train[l] = (r2_train_ridge)
    resultados_mpg_test[l]=(r2_test_ridge)
    print(f"Lambda: {l}, R2 Entrenamiento: {r2_train_ridge}, R2 Prueba: {r2_test_ridge}")
    
```

Dado que estamos buscando el modelo que mejor generalice, estamos buscando un r2 alto en la prueba porque refleja la capacidad de predecir sobre nuevos datos. Un R2 alto en entrenamiento solo es útil si también se mantiene alto en prueba, de lo contrario sería un overfit. Lo ideal es un R2 decente en entrenamiento y lo más alto posible en prueba, manteniendo ambos relativamente cercanos.Considero que la mejor penalización fue la de lambda= 1, ya que explica el 94%% del modelo de entrenamiento y el 60% del modelo de prueba. La de lambda 0.1 también es buena e incluso tiene una mayor explicación sobre el modelo de entrenamiento pero menor sobre la de prueba, lo cual no es muy confiable, considerando que puede haber overfitting.
